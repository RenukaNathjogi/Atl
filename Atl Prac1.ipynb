```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = sns.load_dataset("tips")
df.head()
```

```python
sns.boxplot(x=df['size'])
```

```python
sns.boxplot(x=df['total_bill'])
```

```python
sns.boxplot(x='sex', y='total_bill', data=df)
```

```python
sns.boxplot(x='day', y='total_bill', data=df)
```

```python
sns.boxplot(x='day', y='total_bill', data=df, hue='smoker')
```

```python
sns.boxplot(x='day', y='total_bill', data=df, hue='time')
```

```python
sns.boxplot(x='day', y='total_bill', data=df, order=['Sun' , 'Fri' , 'Thur' , 'Sat'])
```

```python
sns.boxplot(x='sex', y='tip', data=df, order=['Female', 'Male'])
```
